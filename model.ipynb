```python
import pandas as pd
from transformers import XLMRobertaTokenizer, XLMRobertaModel
import torch
from transformers import BartForConditionalGeneration, BartTokenizer
```

```python
df = pd.read_csv(r'C:\Users\qubadlim\Desktop\metbuat.az scraping\metbuataz.csv')
```

```python
model_name = "xlm-roberta-base"
tokenizer = XLMRobertaTokenizer.from_pretrained(model_name)
model = XLMRobertaModel.from_pretrained(model_name)
```

```python
summarizer = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')
summarizer_tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')
```

```python
def summarize_text(text):
    inputs = summarizer_tokenizer([text], max_length=1024, return_tensors='pt', truncation=True)
    summary_ids = summarizer.generate(inputs['input_ids'], num_beams=4, max_length=100, early_stopping=True)
    return summarizer_tokenizer.decode(summary_ids[0], skip_special_tokens=True)
```

```python
for index, row in df.iterrows():
    haber_icerigi = row["content"]
    ozet = summarize_text(haber_icerigi)
    df.at[index, "summary"] = ozet

df.to_csv("guncellenmis_veri_seti.csv", index=False)
```
